# Ridge Regression class from **Scratch**

```python
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt
```

```python
x, y = make_regression(n_samples=160, n_features=1, n_informative=1, n_targets=1, noise=25, random_state=9)
```

```python
plt.scatter(x, y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x, y)
```

```python
print(f'Slope(alpha=0): {lr.coef_}')
print(f'Intercept(alpha=0): {lr.intercept_}')
```

```python
from sklearn.linear_model import Ridge
```

```python
rr = Ridge(alpha=10)
```

```python
rr.fit(x, y)
```

```python
print(f'Slope(alpha=10): {rr.coef_}')
print(f'Intercept(alpha=10): {rr.intercept_}')
```

```python
rr1 = Ridge(alpha=100)
rr1.fit(x, y)
print(f'Slope(alpha=100): {rr1.coef_}')
print(f'Intercept(alpha=100): {rr1.intercept_}')
```

```python
plt.figure(figsize=(6,4))
plt.plot(x, y, 'b.')
plt.plot(x, lr.predict(x), 'r', label='alpha=0')
plt.plot(x, rr.predict(x), 'g', label='alpha=10')
plt.plot(x, rr1.predict(x), 'y', label='alpha=100')
plt.legend()
plt.show()
```

### Creating class for Ridge Regularization from scratch

```python
class RidgeRegularization:
    def __init__(self, alpha=0):
        self.intercept_ = None
        self.coef_ = None
        self.alpha = alpha

    def fit(self, x_train, y_train):
        num = 0
        den = 0
        for i in range(x_train.shape[0]):
            num += (x_train[i]-np.mean(x_train)) * (y_train[i]-np.mean(y_train))
            den += (x_train[i]-np.mean(x_train))**2
            
        self.coef_ = num/(den+self.alpha)
        
        self.intercept_ = np.mean(y_train) - self.coef_*np.mean(x_train)

    def predict(self, x_test):
        return self.coef_*x_test - self.intercept_
```

```python
rdg_self = RidgeRegularization(alpha=10)
rdg_self.fit(x, y)
```
