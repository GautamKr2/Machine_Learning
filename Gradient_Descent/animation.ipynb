# Animation for evaluation of value of b (intercept) on each epoches

```python
import numpy as np
from sklearn.datasets import make_regression

%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import matplotlib.animation as animation
```

```python
x, y = make_regression(n_samples=100, n_features=1, n_informative=1, n_targets=1, noise=20, random_state=4)
```

```python
plt.figure(figsize=(5,3.5))
plt.scatter(x, y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x, y)
print(reg.coef_)
print(reg.intercept_)
```

```python
b = 100
m = 74.9335
lr = 0.001
all_b = []
all_cost = []

epochs = 30
plt.figure(figsize=(7,5))

for i in range(epochs):
    slope = 0
    cost = 0
    for j in range(x.shape[0]):
        slope = slope - 2*(y[j] - (m * x[j]) -  b)
        cost = cost + (y[j] - m * x[j] -b) ** 2

    b = b - (lr * slope)
    all_b.append(b)
    all_cost.append(cost)
    y_pred = m * x + b
    plt.plot(x, y_pred)

plt.scatter(x, y)
```

```python
all_b = np.array(all_b).ravel()
all_b
```

```python
all_cost = np.array(all_cost).ravel()
all_cost
```
