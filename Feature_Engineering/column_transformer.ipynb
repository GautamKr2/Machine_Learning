# Column Transformer in Machine Learning

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('E:\Dataset\covid_toy.csv')
df.sample(5)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test, y_train,y_test = train_test_split(df.drop(columns=['has_covid']), df['has_covid'], test_size=0.2, random_state=1)
```

```python
x_train.shape, x_test.shape
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

## Simple method of Transformation

```python
# Adding imputer on fever column
si = SimpleImputer()
x_train_fever = si.fit_transform(x_train[['fever']])  # Numpy array

x_test_fever = si.transform(x_test[['fever']])
x_train_fever.shape
```

```python
# Adding OneHotEncoder on cough column
oe = OrdinalEncoder(categories=[['Mild','Strong']])

x_train_cough = oe.fit_transform(x_train[['cough']])  # Columns are in numerical form

x_test_cough = oe.fit_transform(x_test[['cough']])
x_train_cough.shape
```

```python
# Adding OneHotEncoder on gender, and city column
ohe = OneHotEncoder(sparse_output=False, drop='first')

x_train_gender_city = ohe.fit_transform(x_train[['gender', 'city']])

x_test_gender_city = ohe.transform(x_test[['gender', 'city']])
x_train_gender_city.shape
```

```python
x_train_age = x_train.drop(columns=['gender','fever','cough','city']).values  # numpy array
x_test_age = x_test.drop(columns=['gender','fever','cough','city']).values
x_train_age.shape
```

```python
x_train_transformed = np.concatenate((x_train_age, x_train_gender_city, x_train_fever, x_train_cough), axis=1)
x_test_transformed = np.concatenate((x_test_age, x_test_gender_city, x_test_fever, x_test_cough),axis=1)
x_train_transformed.shape
```

```python
x_train_transformed = pd.DataFrame(x_train_transformed)
x_train_transformed.head()
```
