```python
import numpy as np
import pandas as pd
```

### Extracting title of Name

```python
df['Name'].str.split(',', expand=True).sample(3)
```

```python
df['Name'].str.split(',', expand=True)[1].str.split('.', expand=True).sample(3)
```

```python
df['title'] = df['Name'].str.split(',', expand=True)[1].str.split('.', expand=True)[0]
```

```python
df[['title', 'Name']].head()
```

```python
(df.groupby('title')['Survived'].mean()).sort_values(ascending=False)
```
