### Applying Feature Construction

```python
x1 = x.copy()
```

```python
x1.sample(2)
```

```python
x1['family'] = x1['SibSp'] + x1['Parch'] + 1
```

```python
x1.sample()
```

#### Function for defining family type

```python
def family_size(num):
    if num==1:
        # Alone
        return 0
    elif num>1 and num<=4:
        # Small Family
        return 1
    else:
        # Large Family
        return 2
```

```python
x1['family_type'] = x1['family'].apply(family_size)
```

```python
x1.sample(2)
```

```python
x1.drop(columns=['SibSp','Parch','family'], inplace=True)
```

```python
x1.sample(2)
```

```python
cross_val_score(LogisticRegression(), x1, y, scoring='accuracy', cv=20).mean()
```
