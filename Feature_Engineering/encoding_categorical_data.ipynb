# Encoding Categorical data

## a) Ordinal Encoding (For input cols), Label encoding (For output col)

```python
import numpy as np
import pandas as pd
```

### Loading dataset

```python
df = pd.read_csv('E:/Dataset/customer.csv')
```

```python
df.sample(5)
```

```python
df = df.iloc[:,2:]
```

```python
df.head()
```

### Working on ordinal categorical column

#### Input columns (Ordinal encoding)

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.iloc[:,0:2], df.iloc[:, -1], test_size=0.2, random_state=2)
```

```python
x_train.shape, x_test.shape
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=[['Poor','Average','Good'], ['School','UG','PG']])
```

```python
oe.fit(x_train)
```

```python
x_train_scaled = oe.transform(x_train)
x_test_scaled = oe.transform(x_test)
```

```python
x_train_scaled
```

```python
oe.categories_
```

#### Output column (Label encoding)

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
le.fit(y_train)
```

```python
le.classes_
```

```python
y_train_scaled = le.transform(y_train)
y_test_scaled = le.transform(y_test)
```

```python
y_train_scaled
```
