# Iterative imputer or MICE(Multivariate Imputation by Chained Equations)

```python
import numpy as np
import pandas as pd

from sklearn.linear_model import LinearRegression
```

### Doing data ready to understand this concept

```python
df = np.round(pd.read_csv('E:/Dataset/50_startups.csv')/10000)
```

```python
np.random.seed(4)
df = df.sample(5).copy()
df
```

```python
df = df.iloc[:,:-1].copy()
df
```

```python
df.iloc[1,0] = np.nan
df.iloc[3,1] = np.nan
df.iloc[4,2] = np.nan
```

```python
df
```

## Iteration - 1

### Step 1 - Impute all missing values with mean of respective column

```python
df0 = pd.DataFrame()

df0['R&D Spend'] = df['R&D Spend'].fillna(df['R&D Spend'].mean())
df0['Administration'] = df['Administration'].fillna(df['Administration'].mean())
df0['Marketing Spend'] = df['Marketing Spend'].fillna(df['Marketing Spend'].mean())
```

```python
# 0th Iteration
df0
```

## Iteration - 2

### Step 2 - Making each columns' NaN value again as NaN one by one

```python
df1 = df0.copy()

df1.iloc[1,0] = np.nan

df1
```

```python
# Use rows which have not missing value as input

x = df1.iloc[[0,2,3,4],1:3]
x
```

```python
y = df1.iloc[[0,2,3,4],0]
y
```

```python
lr = LinearRegression()
lr.fit(x,y)
lr.predict(df1.iloc[1,1:].values.reshape(1,2))
```

```python
df1.iloc[1,0] = 22.16
```

```python
df1
```

```python
df1.iloc[3,1] = np.nan
df1
```

```python
x = df1.iloc[[0,1,2,4],[0,2]]
x
```

```python
y = df1.iloc[[0,1,2,4],1]
y
```

```python
lr = LinearRegression()
lr.fit(x,y)
lr.predict(df1.iloc[3,[0,2]].values.reshape(1,2))
```

```python
df1.iloc[3,1] = 10.18
df1
```

```python
df1.iloc[4,2] = np.nan
df1
```

```python
x = df1.iloc[:4,:-1]
x
```

```python
y = df1.iloc[:4,-1]
y
```

```python
lr = LinearRegression()
lr.fit(x,y)
lr.predict(df1.iloc[4,:-1].values.reshape(1,2))
```

## Iteration - 3

#### - Now Repeat the above steps 2, and 3 till subtraction of two consecutive iteration df is zero or tend to zero

```python
df2 = df1.copy()
df2.iloc[1,0] = np.nan
df2
```

```python
x = df2.iloc[[0,2,3,4], 1:]
y = df2.iloc[[0,2,3,4], 0]
lr = LinearRegression()
lr.fit(x, y)
lr.predict(df2.iloc[1,1:].values.reshape(1,2))
```

```python
df2.iloc[1,0] = 22.34
```

```python
df2.iloc[3,1] = np.nan
x = df2.iloc[[0,1,2,4], [0,2]]
y = df2.iloc[[0,1,2,4], 1]

lr = LinearRegression()
lr.fit(x,y)
lr.predict(df2.iloc[3, [0,2]].values.reshape(1,2))
```

```python
df2.iloc[3,1] = 10.22
```

```python
df2.iloc[4,-1] = np.nan
x = df2.iloc[:-1, :-1]
y = df2.iloc[:-1, -1]

lr = LinearRegression()
lr.fit(x,y)
lr.predict(df2.iloc[4,:-1].values.reshape(1,2))
```

```python
df2.iloc[-1,-1] = 30.16
```

```python
df2
```

```python
df2 - df1
```

## Iteration - 4

```python
df3 = df2.copy()

df3.iloc[1,0] = np.nan
x = df3.iloc[[0,2,3,4], 1:]
y = df3.iloc[[0,2,3,4], 0]
lr = LinearRegression()
lr.fit(x, y)
pred = lr.predict(df3.iloc[1,1:].values.reshape(1,2))
df3.iloc[1,0] = np.round(pred,2)

df3.iloc[3,1] = np.nan
x = df3.iloc[[0,1,2,4], [0,2]]
y = df3.iloc[[0,1,2,4], 1]
lr = LinearRegression()
lr.fit(x,y)
pred = lr.predict(df3.iloc[3, [0,2]].values.reshape(1,2))
df3.iloc[3,1] = np.round(pred,2)

df3.iloc[4,-1] = np.nan
x = df3.iloc[:-1, :-1]
y = df3.iloc[:-1, -1]
lr = LinearRegression()
lr.fit(x,y)
pred = lr.predict(df3.iloc[4,:-1].values.reshape(1,2))
df3.iloc[4,-1] = np.round(pred,2)
```

```python
df3
```
