# Imputation using Missing Indicator

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Loading dataset and splitting

```python
df = pd.read_csv('E:/Dataset/titanic-Dataset.csv', usecols=['Age', 'Fare', 'Survived'])
```

```python
df.sample(2)
```

```python
x = df.drop(columns=['Survived'])
y = df['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

### Calculating accuracy_score on imputing using SimpleImputer

```python
from sklearn.impute import SimpleImputer
```

```python
si = SimpleImputer()
x_train_trf = si.fit_transform(x_train)
x_test_trf = si.fit_transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logReg = LogisticRegression()
logReg.fit(x_train_trf, y_train)
y_pred = logReg.predict(x_test_trf)

from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

### Calculating accuracy_score on imputation using Missing Indicator

```python
from sklearn.impute import MissingIndicator
```

```python
misInd = MissingIndicator()
x_train_miss = misInd.fit_transform(x_train)
x_test_miss = misInd.transform(x_test)
```

```python
x_test_miss
```

```python
misInd.features_
```

```python
x_train['age_NaN'] = x_train_miss
x_test['age_NaN'] = x_test_miss
```

```python
x_test.sample(4)
```

```python
si = SimpleImputer()
x_train_trf2 = si.fit_transform(x_train)
x_test_trf2 = si.transform(x_test)
```

```python
logReg.fit(x_train_trf2, y_train)
y_pred2 = logReg.predict(x_test_trf2)

accuracy_score(y_test, y_pred2)
```
