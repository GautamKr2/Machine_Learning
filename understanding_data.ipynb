## III. Perform EDA using Bivariate, and Multivariate Analysis

### Importing libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Loading all required dataset

```python
tips = sns.load_dataset('tips')
```

```python
tips.sample(5)
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.sample(5)
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.sample(5)
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.sample(5)
```

### 1. Scatter Plot (Numerical-Numerical)

#### - Bivariate

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x=tips['total_bill'], y=tips['tip'])
```

#### - Multivariate

```python
plt.figure(figsize=(8,7))
sns.scatterplot(x=tips['total_bill'], y=tips['tip'], hue=tips['sex'], style=tips['smoker'], size=tips['size'])
```

### 2. Bar Plot (Numerical-Categorical)

#### - Bivariate

```python
plt.figure(figsize=(4,3))
sns.barplot(x=titanic['pclass'], y=titanic['age'])
```

#### - Multivariate

```python
#plt.figure(figsize=(4,3))
sns.barplot(x=titanic['pclass'], y=titanic['age'], hue=titanic['sex'])
```

### 3. Box Plot (Numerical-Categorical)

#### - Bivariate

```python
plt.figure(figsize=(6,4))
sns.boxplot(y=titanic['sex'], x=titanic['age'])
```

#### - Multivariate

```python
sns.boxplot(x=titanic['age'], y=titanic['sex'], hue=titanic['survived'])
```

### 4. Hist Plot with KDE(Kernel Density Function) (Numerical-Categorical)

```python
sns.histplot(titanic['age'], kde=True)
```

```python
sns.kdeplot(titanic[titanic['survived']==0]['age'], label="not survived")
sns.kdeplot(titanic[titanic['survived']==1]['age'], label="survived")
plt.legend()
plt.figtext(0.65, -0.025, "- ye graph bata raha hai ki ek age ke logon ka marne ya bachne ka probability kya hai", ha='center')
```

### 5. Heat Map (Categorical-Categorical)

```python
pd.crosstab(titanic['pclass'], titanic['survived'])
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(pd.crosstab(titanic['pclass'], titanic['survived']))
```

```python
titanic.groupby('pclass')['survived'].mean()*100
```

```python
plt.figure(figsize=(5,4))
(titanic.groupby('pclass')['survived'].mean()*100).plot(kind='bar')
```

### 6. Cluster Map (Categorical-Categorical)

```python
pd.crosstab(titanic['sibsp'], titanic['survived'])
```

```python
sns.clustermap(pd.crosstab(titanic['parch'], titanic['survived']))
```

### 7. Pair Plot (Will give scatter plot between all the numerical columns with each other)

```python
iris.head()
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue='species')
```

### 8. Line Plot (Numerical-Numerical)

```python
flights.head()
```

```python
grouped_pass = flights.groupby('year').sum(numeric_only=True).reset_index()
grouped_pass
```

```python
sns.lineplot(x=grouped_pass['year'], y=grouped_pass['passengers'])
```

### Designing Heat Map, and Cluster Map on flights dataset

```python
flights.pivot(values='passengers', index='month', columns='year')
```

```python
sns.heatmap(flights.pivot(values='passengers', index='month', columns='year'))
```

```python
sns.clustermap(flights.pivot(values='passengers', index='month', columns='year'))
```
