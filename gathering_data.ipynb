#### Work for a single page

```python
companies = soup.find_all('div', class_='companyCardWrapper')
```

```python
len(companies)
```

```python
name = []
rating = []
reviews = []
c_type = []
loc = []
salary = []

i = 0

for company in companies:
    name.append(company.find('h2').text.strip())
    rating.append(company.find('div', class_='rating_text').text.strip())
    rat_count = company.find('span', class_='companyCardWrapper__companyRatingCount').text.strip()
    reviews.append(company.find('span', class_='companyCardWrapper__ActionCount').text)
    c_type.append(company.find('span', class_='companyCardWrapper__interLinking').text.strip().split('|')[0])
    loc.append(company.find('span', class_='companyCardWrapper__interLinking').text.strip().split('|')[1])
    salary.append(company.find('span', class_='companyCardWrapper__ActionCount').text)
    rating[i] = rating[i]+" "+rat_count
    i += 1
```

```python
d = {'comp_name':name, 'comp_rating':rating, 'comp_reviews':reviews, 'comp_type':c_type, 'location':loc, 'salaries':salary}
```

#### Working for multiple pages

```python
company_dfs = pd.DataFrame() # Created a empty dataframe for the whole webpage

for i in range(0,100):
    url = f"https://www.ambitionbox.com/list-of-companies?page={i+1}"
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"}
    webpage = requests.get(url, headers=headers).text
    soup = BeautifulSoup(webpage, 'lxml')
    companies = soup.find_all('div', class_='companyCardWrapper')
    
    name = []
    rating = []
    reviews = []
    c_type = []
    loc = []
    salary = []
    i = 0

    for company in companies:
        name.append(company.find('h2').text.strip())
        rating.append(company.find('div', class_='rating_text').text.strip())
        rat_count = company.find('span', class_='companyCardWrapper__companyRatingCount').text.strip()
        reviews.append(company.find('span', class_='companyCardWrapper__ActionCount').text)
        c_type.append(company.find('span', class_='companyCardWrapper__interLinking').text.strip().split('|')[0])
        try:
            loc.append(company.find('span', class_='companyCardWrapper__interLinking').text.strip().split('|')[1])
        except:
            loc.append(pd.NA)
        salary.append(company.find('span', class_='companyCardWrapper__ActionCount').text)
        rating[i] = rating[i]+" "+rat_count
        i += 1

    d = {'comp_name':name, 'comp_rating':rating, 'comp_reviews':reviews, 'comp_type':c_type, 'location':loc, 'salaries':salary}
    df = pd.DataFrame(d)

    company_dfs = pd.concat([company_dfs, df], ignore_index=True)
```
